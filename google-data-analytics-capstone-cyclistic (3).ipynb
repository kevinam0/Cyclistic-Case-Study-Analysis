**DATA CLEANING AND WRANGLING BY VARIABLE**

**ride_id**

Cleaning - Remove duplicates to ensure each ride has one unique, corresponding observation/row

```r
CyclisticMergedNoDup <- CyclisticMerged %>% 
distinct(ride_id, .keep_all = TRUE)

print(paste("Removed", nrow(CyclisticMerged) - nrow(CyclisticMergedNoDup),
            "duplicate rows corresponding to ride_id"))
```

Cleaning - check for and remove any missing values (NA)

```r
print(paste("ride_id has", sum(is.na(CyclisticMergedNoDup$ride_id)), "missing values" ))
```

**rideable_type** 

Transformation - convert data type from character to factor for an easier analysis later on

```r
CyclisticMergedNoDup$rideable_type <- as.factor(CyclisticMergedNoDup$rideable_type)

class(CyclisticMergedNoDup$rideable_type)
nlevels(CyclisticMergedNoDup$rideable_type)
summary(CyclisticMergedNoDup$rideable_type)
```

docked_bike has been discovered as a level within the rideable_type variable. It only occurs in 2.25% of all rides. For more clarity and information on this type of bike, I contacted Divvy's data support team via email.

They clarified that classic bikes used to be recorded as docked_bikes up until April 2020, as classic bikes must dock at a station to end a ride. On the other hand, electric bikes can be docked or alternatively locked outside of stations. After April 2020, Divvy started recording classic bikes as classic_bikes instead of docked_bikes. 

With this information, let's make the assumption that docked_bike refers to Divvy's classic bike, and change their values accordingly to classic_bike.

```r
CyclisticMergedNoDup$rideable_type[CyclisticMergedNoDup$rideable_type == "docked_bike"] <- "classic_bike"

summary(CyclisticMergedNoDup$rideable_type)
```

**started_at, ended_at** 

Both columns are a dttm data type which contain the many components of ride time data that can be extracted into their own variables and analyzed/visualized seperately 

Transformation - create new column of calculated ride time by using the difftime function to find the difference between end time and start time of a ride

```r
CyclisticMergedNoDup$ride_time_minutes <- (as.double(difftime(CyclisticMerged$ended_at, CyclisticMerged$started_at))) /60

summary(CyclisticMergedNoDup$ride_time_minutes)
```

Cleaning

From viewing summary statistics of the newly created column, it can be seen that there are obvious outliers that should be removed in order to perform an accurate analysis later on.

1. Negative ride times - doesn't make any logical sense
2. Abnormally long rides 
3. Rides under 60 seconds - classified by divvy as potential false starts or attempts at redocking, source:
   [https://divvybikes.com/system-data](http://divvybikes.com/system-data)

Seperate ride_time_minutes into quantiles to see how ride time is distrubuted and then how to filter values

```r
quantile_ride_time = quantile(CyclisticMergedNoDup$ride_time_minutes, seq(0, 1, by=0.01))
quantile_ride_time
```

```r
CyclisticMergedNoDup <- CyclisticMergedNoDup %>% 
  filter(ride_time_minutes >=  1) %>%
  filter(ride_time_minutes <= 180) # 3 hours, exlude large ride times to not skew mean

summary(CyclisticMergedNoDup$ride_time_minutes)
```

Parsing - Extracting the month, weekday, and hour of each ride from started_at into their own respective variables, using strftime and floor_date functions from lubridate package

```r
CyclisticMergedNoDup <- CyclisticMergedNoDup %>% 
  mutate(month = strftime(CyclisticMergedNoDup$started_at, "%B"),
         weekday = strftime(CyclisticMergedNoDup$started_at, "%A"))
        
CyclisticMergedNoDup <- CyclisticMergedNoDup %>% 
  mutate(hour_started = floor_date(started_at, unit = "hour")) 
CyclisticMergedNoDup <- CyclisticMergedNoDup %>% 
  mutate(hour_started = strftime(CyclisticMergedNoDup$hour_started,"%R"))

head(CyclisticMergedNoDup)
```

Transformation - Change data type of new columns from character to factor in order for an easier analysis later on

```r
CyclisticMergedNoDup$month <- as.factor(CyclisticMergedNoDup$month)                          
CyclisticMergedNoDup$weekday <- as.factor(CyclisticMergedNoDup$weekday)                          
CyclisticMergedNoDup$hour_started <- as.factor(CyclisticMergedNoDup$hour_started)

class(CyclisticMergedNoDup$month)
nlevels(CyclisticMergedNoDup$month)
summary(CyclisticMergedNoDup$month)
class(CyclisticMergedNoDup$weekday)
nlevels(CyclisticMergedNoDup$weekday)
summary(CyclisticMergedNoDup$weekday)
class(CyclisticMergedNoDup$hour_started)
nlevels(CyclisticMergedNoDup$hour_started)
summary(CyclisticMergedNoDup$hour_started)
```

**start_station_name, end_station_name, start_station_id, end_station_id**

Transformation - Change data types of columns from character to factor

```r
CyclisticMergedNoDup$start_station_name <- as.factor(CyclisticMergedNoDup$start_station_name)                          
CyclisticMergedNoDup$end_station_name <- as.factor(CyclisticMergedNoDup$end_station_name)                          
CyclisticMergedNoDup$start_station_id <- as.factor(CyclisticMergedNoDup$start_station_id)                          
CyclisticMergedNoDup$end_station_id <- as.factor(CyclisticMergedNoDup$end_station_id)

class(CyclisticMergedNoDup$start_station_name)
nlevels(CyclisticMergedNoDup$start_station_name)
class(CyclisticMergedNoDup$end_station_name)
nlevels(CyclisticMergedNoDup$end_station_name)
class(CyclisticMergedNoDup$start_station_id)
nlevels(CyclisticMergedNoDup$start_station_id)
class(CyclisticMergedNoDup$end_station_id)
nlevels(CyclisticMergedNoDup$end_station_id)
```

**start_lat, start_lng, end_lat, end_lng**

Transformation - Create column of calculated distance between start and end points with the coordinate values contained in these 4 variables, use distGeo function

```r
CyclisticMergedNoDup$distance_between_stations_km <- 
  distGeo(matrix(c(CyclisticMergedNoDup$start_lng,CyclisticMergedNoDup$start_lat), ncol = 2),
          matrix(c(CyclisticMergedNoDup$end_lng,CyclisticMergedNoDup$end_lat), ncol = 2)) / 1000

summary(CyclisticMergedNoDup$distance_between_stations_km)
```

Cleaning - outliers, NA's

Distances with a value of 0 indicate riders that ended their rides at the same geopoint they started at (round trip) and should be kept to be analyzed further

Distances with very high values, clearly identified as outliers and should be further examined and removed before performing analysis on said variable

444 missing values identified - observations can be kept to perform analysis on other variables, subsequentially can be removed when performing specific analysis relating to distance

```r
CyclisticMergedNoDup <- CyclisticMergedNoDup %>%
  arrange(desc(distance_between_stations_km))

head(CyclisticMergedNoDup$distance_between_stations_km, n = 10)
```

After arranging the data frame by distance, sorted from largest to smallest values, 8 outliers with illogical values have been identified to be removed

```r
CyclisticMergedNoDup <- CyclisticMergedNoDup %>% 
slice(-(1:8))

head(CyclisticMergedNoDup$distance_between_stations_km, n = 10)
summary(CyclisticMergedNoDup$distance_between_stations_km)
```

# ANALYZE

Throughout our analysis process, we will be using the useful functions from the dplyr package to group and aggregate the cleaned data to our choosing. Then to visualize we'll use the ggplot package. 

To start off let's summarise the data by the number of rides taken by each User type to see how they compare.

```r
member_casual_summary <- CyclisticClean %>% 
  group_by(member_casual) %>% 
  summarise(count = n(),  percentage = round(length(ride_id)/nrow(CyclisticClean)*100,2), 
            .groups = "drop")

options(scipen = 999, repr.plot.width = 11, repr.plot.height = 8)

ggplot(member_casual_summary, aes(x = " " , y = count, fill = member_casual)) + 
  geom_col() + labs(title = "Number of Rides Taken by User Type (Aug 2022 - July 2023)") + 
  geom_text(aes(label = paste(count," --- ", percentage ,"%")), position = position_stack(vjust = 0.5), 
  size = 6) + theme_void(base_size = 19) + scale_fill_discrete(name = "User Type") + 
  coord_polar(theta = "y")
```

There is a significant difference in the amount of rides taken between these 2 customer groups ; roughly 1.35 million!

Annual members are the clear majority of Cyclistic ride takers with **62%** , while casual riders comprise **38%** of total rides taken during the measured annual period.

Let's take a deeper look into the differences in ride count by grouping the counts by the Month that the rides took place.

```r
CyclisticClean$month <- ordered(CyclisticClean$month, levels = 
  c("August", "September", "October", "November", "December", "January",
  "February", "March", "April", "May", "June", "July"))

month_member_casual_summary <- CyclisticClean %>% 
  group_by(month, member_casual) %>%
  summarise(count = n(), .groups = "drop")%>%
  arrange(month, member_casual)

options(repr.plot.width = 16, repr.plot.height = 8)

ggplot(data = month_member_casual_summary, aes(x = month, y = count, fill = member_casual),
  show.legend = FALSE) + geom_col(position = "dodge", width = 0.6) +
  labs(title = "Monthly Ride Count by User Type (Aug 2022 - July 2023)", x = "Month", y = "# of Rides") + 
  scale_fill_discrete(name="User Type") + theme_gray(base_size = 19)
```

A number of clear trends can be identified from the above plot:

1. The number of rides for both User types has a positive correlation and probable causation relating to    the month and it's respective seasonal weather/conditions in the City of Chicago. The months where      temperatures are at their highest (June, July, August, September), are when the most amount of rides    take place. Conversely, in the winter months where temperatures are low and there is snow as            precipitation, ride count drops significantly.

2. As the City departs from the winter, the ride count of casual users starts to catch up to annual        members, month-by-month until August where the counts are closest in relation to each other. 
  
 Now let's group ride counts by the days of the week.

```r
CyclisticClean$weekday <- ordered(CyclisticClean$weekday, levels = 
                                        c("Monday", "Tuesday", "Wednesday", "Thursday", 
                                          "Friday", "Saturday", "Sunday"))

weekday_member_casual_summary <- CyclisticClean %>% 
  group_by(weekday, member_casual) %>%
  summarise(count = n(), .groups = "drop")%>%
  arrange(weekday, member_casual)

ggplot(data = weekday_member_casual_summary, aes(x = weekday, y = count, fill = member_casual),
  show.legend = FALSE) + geom_col(position = "dodge", width = 0.8) +
  labs(title = "Ride Count by Day of the Week (Aug 2022 - July 2023)", x = "Day", y = "# of Rides") + 
  scale_fill_discrete(name="User Type") + theme_gray(base_size = 19) + geom_text(aes(label = count),
  size = 5, vjust = -0.25, position = position_dodge(width = 0.8))
```

During the traditional work week (Monday to Friday) annual members are the clear majority of ride takers. However, the gap is significantly less during the weekend.

Perhaps, annual members are using their rides for commuting to work and other scheduled events that would take place during the work week. On the other hand, many casual riders may use Cyclistic bikes for more leisurely purposes (tourism, exercise, etc.) that would more likely take place during weekends. 

Let's take a deeper dive by looking at the time of day these rides are taking place for both the workweek and weekend. 

```r
hour_member_casual_summary <- CyclisticClean %>% 
  filter(weekday == "Monday" | weekday == "Tuesday"| weekday == "Wednesday" |
           weekday == "Thursday" |weekday == "Friday") %>% 
  group_by(hour_started, member_casual) %>%
  summarise(count = n(), .groups = "drop")%>%
  arrange(hour_started, member_casual)

ggplot(data = hour_member_casual_summary, aes(x = hour_started, y = count, group = member_casual, 
  color = member_casual)) + geom_point() + geom_line() +
  labs(title = "Ride Count by Time of Day on Weekdays (Aug 2022 - July 2023)", x = "Hour", 
  y = "# of Rides") + guides(color = guide_legend(title = "User Type")) + 
  theme(text = element_text(size = 19),axis.text.x = element_text(angle = 90))

hour_member_casual_summary1 <- CyclisticClean %>% 
  filter(weekday == "Saturday" | weekday == "Sunday") %>% 
  group_by(hour_started, member_casual) %>%
  summarise(count = n(), .groups = "drop") %>%
  arrange(hour_started, member_casual)

ggplot(data = hour_member_casual_summary1, aes(x = hour_started, y = count, group = member_casual,
  color = member_casual)) + geom_point() + geom_line() +
  labs(title = "Ride Count by Time of Day on Weekends (Aug 2022 - July 2023)", x = "Hour",
  y = "# of Rides") + guides(color = guide_legend(title = "User Type")) + 
  theme(text = element_text(size = 19),axis.text.x = element_text(angle = 90))
```

Let's start analyzing the distance between users starting and ending positions. To clarify this doesn't equate to total distance travelled during a ride, as there is no data available to track where users travelled between their starting and ending point. This value is a straight line measurement between the two recorded geopoints. It can be used to better understand trends relating to distance but isn't an exact measurement of distance travelled.

Rides where distance is 0 km indicates a trip where the user returned to their starting position (round trip). Let's subset our data to where this occurs and visualize to explore further.

As electric bikes have recorded coordinates with 2 decimal places (1.1 km precision), let's exclude them from this analysis as it can't be certain that users are indeed returning to the same location they started. As classic bikes must be docked, observations where their distance is 0 ensures that these riders indeed completed a round trip.

```r
distance_summary <- CyclisticClean %>% 
  drop_na(distance_between_stations_km) 

distance0 <- distance_summary %>% 
  filter(distance_between_stations_km == 0) %>% 
  filter(rideable_type == "classic_bike")

dist0_summary <- distance0 %>%  
group_by(member_casual) %>% 
  summarise(count = n(), percentage = round(length(ride_id)/nrow(distance0)*100,2), 
            .groups = "drop")

options(repr.plot.width = 10, repr.plot.height = 8)

ggplot(dist0_summary, aes(x= " ", y = count, fill = member_casual)) +
  geom_col() + coord_polar(theta = "y") + theme_void(base_size = 19) +
  geom_text(aes(label = paste(count, " --- ", percentage,"%")),
  position = position_stack(vjust = 0.5), size = 6) + 
  labs(title ="Count of Round Trips for Classic Bikes by User Type", 
  subtitle = "(Aug 2022 - July 2023)") + scale_fill_discrete(name = "User Type") 

distance0casual <- distance0 %>% 
  filter(member_casual == "casual")

summary(distance0$ride_time_minutes)
summary(distance0casual$ride_time_minutes)
```

**Annual Member Profile:**

As these users hold an annual membership with Cyclistic, it can be assumed that they reside in Chicago or its greater metropolitan area. This group tends to use Cyclistic bikes as a means for getting around the city, taking rides associated with daily living within a city such as Chicago. During the week, users take Cyclistic bikes to commute to/from work at high volumes, noticebly during popular workweek commute times (8:00 am and 5:00pm). When not commuting to work, they take rides for other city living events; perhaps going to an appointment, buying groceries, meeting friends etc. They use our 2 bike types at equal counts, with electic bikes being favored for an efficient ride while classic bikes are favored for a more liesurely ride. Although this group has access to Cyclistic bikes annually, they take significantly more rides during months where temperatures are warm and conditions are suitable for bike riding. 

**Casual User Profile:**

As these users gain access to Cyclistic bikes through either a single ride pass or a day pass, their purpose/tendancies for using our bikes are often different than annual members. 

The seasonality of Chicago has an enormous impact on the # of single ride/day passes being bought. Many of these casual users are tourists that visit the city during times when the temperature is warm (most popular -- July, August, June) and use our bikes as a means for exploring Chicago. 

An influx of casual users occur during weekends and Fridays. Beside tourists, many of these users are residents of the Chicago Metropolitan Area, using Cyclistic bikes for lesiure/exercise in their free time away from weekly events such as work, school, etc. They tend to log longer ride times and more round trips than annual members, once again pointing to using our bikes for exercise. Diving deeper into where casual users are taking rides, the majority tend to take trips within the downtown core and along Lake Michigan.

During the week is where casual users behave most similar to annual members. During popular commute times they have peaks in ride counts just like annual members however they heavily favor our electric bikes for a fast and efficient travel. 

**TOP 3 RECOMMENDATIONS FROM DATA-DRIVEN INSIGHTS**

1. Target casual users who are behaving most like annual members but have yet to purchase a yearly subscription. The most noteworthy example is during heavy workweek commute times (8:00 am and 5:00 pm) when both groups hit peaks in ride counts. Marketing should create a targeted campaign towards these casual riders using Cyclistic bikes to get to/from work, emphasizing the cost-effectiveness of using a membership as opposed to a single ride/day pass. As casual users heavily prefer using e-bikes during these commute times, marketing should also highlight the lower rates for e-bikes that Cyclistic members get access to. In terms of implementation, we should have ad placements in locations and stations where these workweek commutes are taking place. 

2. Create a marketing campaign targeting casual users who are frequently using Cyclistic bikes for leisure and exercise. These customers are most likely Chicagoans who are taking rides when times/conditions are ideal; weekends when they're off work, months when weather is suitable for riding. Therefore the timing of rolling out this plan is key. It is recommended to start this campaign at the begining of Spring, when the weather is warming up and there are many months ahead of ideal riding conditions. Emphasize the healthy lifestyle that is associated with Cyclistic as a brand. Along with on the ground marketing, use social media for targeted ad placements in order to successfully reach this specific customer segment.

3. As casual users are the group taking significantly longer rides, highlight the lower per minute rates that comes along with having a yearly subscription. Also focus on the 45 minutes of free ride time for classic bikes that annual members get access to, highligting the fiscal incentive (specific to their ride behavior) for these customers to upgrade. Implement this marketing strategy through physical location ads (popular route areas for casual users - downtown/along Lake Michigan) and social media.
